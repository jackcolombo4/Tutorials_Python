#### Add a new column to your DataFrame df containing the values "man" and "woman" at random with a percentage distribution of 70% and 30%, respectively

```python
np.random.seed(12)

# To avoid ValueError: Length of values (50) does not match length of index (51)
#gender_distribution = ["man"] * int(len(df) * 0.7) + ["woman"] * int(len(df) * 0.3)

# Create a new column with random "man" and "woman" values
total_samples = len(df)
num_men = int(total_samples * 0.7)
num_women = total_samples - num_men
gender_distribution = ["man"] * num_men + ["woman"] * num_women
```

```python
len(gender_distribution)
```

```python
np.random.shuffle(gender_distribution)
gender_distribution[:10]
```

```python
df["gender"] = gender_distribution
df.head()
```

```python
filtered_df = df[df['abbrev'].str.startswith('A')]
filtered_df
```

```python
# Index of the row you want to duplicate
row_to_duplicate = 1

# Duplicate the specific row and append it to the DataFrame
row = df.loc[row_to_duplicate]
# To avoid the FutureWarning (frame.append method is deprecated)
#filtered_df = filtered_df.append(row, ignore_index=True)

# Create a DataFrame containing the duplicated row
duplicated_row_df = pd.DataFrame([row] * 1, columns=row.index)

# Concatenate the duplicated row DataFrame with the original DataFrame
filtered_df = pd.concat([filtered_df, duplicated_row_df], ignore_index=True)

filtered_df
```

```python
# create plot, assign to ax specify the legend position
ax = sns.countplot(x = 'abbrev', hue = 'gender', data = filtered_df, palette = 'cubehelix')
plt.title('Abbreviation and gender')
ax.legend(loc='upper right')
plt.show()
```

```python
sns.pointplot(x="speeding", y="alcohol", data=filtered_df)
sns.set_style("dark")
plt.show()
```

### Joint plot:
to display two variables with bivariate and univariate graphs

```python
# The palette parameter doesn't work, it is necessary to use scatter_kws
# Add a linear regression 
sns.set_palette("plasma")
sns.jointplot(x='speeding', y='alcohol', data=df,  kind='reg', joint_kws={'color':'green'})
plt.show()
```

```python
# Display kernel density estimate 
sns.set_palette("coolwarm")
sns.jointplot(x='not_distracted', y='no_previous', data=df, kind='kde')
plt.show()
```

```python
# Display hexagons instead of points in scatter plot
sns.jointplot(x='ins_losses', y='alcohol', data=df, kind='hex')
plt.show()
```

### Regplot:
Regplot can be used to visualize the linear relationship as determined through regression.

```python
# need float to avoid TypeError: can't multiply sequence by non-int of type 'float'
ax = sns.regplot(x="alcohol", y="speeding", data=df)
plt.show()
```

```python
""" x_jitter parameter is used to add a small amount of random noise to the x-values of the data points. """
ax = sns.regplot(x="alcohol", y="speeding", data=df, x_jitter=0.1)
plt.show()
```

```python
sns.lmplot(x="total", y="alcohol", col="gender", hue="gender", data=df, col_wrap=2, height=5)
plt.show()
```

### KDE plot (Kernel Density Estimate):

```python
""" Univariate KDE plot """
# "fill=True" instead of 'shade' is deprecated 
sns.kdeplot(df.loc[(df['gender']=='man'), 
            'alcohol'], color='b', fill=True, label='man') 
sns.kdeplot(df.loc[(df['gender']=='woman'), 
            'alcohol'], color='r', fill=True, label='woman')
plt.show()
```

```python
# Setting up the samples 
df_man = df.query("gender=='man'") 
df_woman = df.query("gender=='woman'") 

df_woman
```

```python
""" Bivariate KDE plot """
# `shade_lowest` has been replaced by `thresh`; setting `thresh=0.05.
sns.kdeplot(data=df_man, x='speeding', y='alcohol', cmap="Reds", fill=True, thresh=0.05, label='man')
sns.kdeplot(data=df_woman, x='speeding', y='alcohol', cmap="Blues", fill=True, thresh=0.05, label='woman')
plt.show()
```

### Heatmaps:

```python
# Get another dataset for trials
df_flights=sns.load_dataset("flights")
df_flights.head()
```

```python
""" Create pivot table based on column values. 
It is a reshaped DataFrame, organized by given index / column values.
=> All arguments of DataFrame.pivot should be keyword-only (avoid FutureWarning)
"""
flights_pivot = df_flights.pivot(index="month", columns="year", values="passengers")
flights_pivot
```

```python
sns.heatmap(flights_pivot, linewidths=.5, cmap="coolwarm")
plt.show()
```

```python
"""  Annot parameter, to add annotation to heatmap.
Annotations are the actual data values from the DataFrame that the heatmap is visualizing. 
(Example with df => car_crash dataset)
"""
# Calculate the correlation matrix
# Specify "numeric_only" since the default value of in DataFrame.corr is deprecated. 
corr = df.corr(numeric_only=True) 
# Create a heatmap with annotations
sns.heatmap(corr, annot=True, linewidths=.5, cmap="YlGnBu")
plt.show()
```

### Cluster map:
Plot a matrix dataset as a hierarchically-clustered heatmap, to order data by similarity. 

```python
sns.clustermap(flights_pivot,linewidths=.5,cmap="magma")
plt.show()
```
