## Classic activation functions

```python
""" Functions """
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def relu(x):
    return np.maximum(0, x)
def tanh(x):
    return np.tanh(x)
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x))
def leaky_relu(x, alpha=0.03):
    return np.where(x >= 0, x, alpha * x)
```

```python
## Set up the plot
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

## Plot the sigmoid (top left subplot)
x = np.linspace(-10, 10, 100)
y = sigmoid(x)
axs[0, 0].plot(x, y, color="navajowhite")
axs[0, 0].set_xlabel("X", fontsize=12)
axs[0, 0].set_ylabel("Y", fontsize=12)
axs[0, 0].set_title("Sigmoid Function", fontsize=16)

## Plot the ReLU (top right subplot)
x = np.linspace(-10, 10, 100)
y = relu(x)
axs[0, 1].plot(x, y, color="mediumseagreen")
axs[0, 1].set_xlabel("X", fontsize=12)
axs[0, 1].set_ylabel("Y", fontsize=12)
axs[0, 1].set_title("ReLU Function", fontsize=16)

## Plot the tanh (bottom left subplot)
x = np.linspace(-10, 10, 100)
y = tanh(x)
axs[1, 0].plot(x, y, color="slategrey")
axs[1, 0].set_xlabel("X", fontsize=12)
axs[1, 0].set_ylabel("Y", fontsize=12)
axs[1, 0].set_title("Tanh Function", fontsize=16)

# Plot the Leaky ReLU function
x_leaky_relu = np.linspace(-10, 10, 100)
y_leaky_relu = leaky_relu(x_leaky_relu)
axs[1, 1].plot(x_leaky_relu, y_leaky_relu, color="orange")
axs[1, 1].set_xlabel("X", fontsize=12)
axs[1, 1].set_ylabel("Y", fontsize=12)
axs[1, 1].set_title("Leaky ReLU Function", fontsize=16)

## Adjust spacing between subplots
plt.subplots_adjust(wspace=0.3, hspace=0.4)
```

```python
fig, ax = plt.subplots(figsize=(4, 3))
## Plot the softmax (bottom right subplot)
x = np.array([-1, 1])
y = softmax(x)

ax.bar(["Class 1", "Class 2"], y, color="steelblue")
ax.set_xlabel("Classes", fontsize=12)
ax.set_ylabel("Output", fontsize=12)
ax.set_title("Softmax Function", fontsize=16)

# Adjust spacing between subplots
plt.subplots_adjust(wspace=0.3, hspace=0.4)
plt.show()
```

## Special case of a radial function: sine of the Euclidean distance from the origin. 
Each point values ranging between -1 and 1, and it will be symmetric about the origin.
It calculates the ratio of the sine of the distance to the distance itself.

Geometrically, this corresponds to a series of concentric circles centered at the origin, where the function value varies based on the distance from the center. 

 
Formula: 

$
f(x, y) = \frac{\sin(\sqrt{x^2 + y^2})}{\sqrt{x^2 + y^2}}
$

```python
plt.style.use(['fivethirtyeight'])

# Define the function to plot
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2)) / np.sqrt(x**2 + y**2)

#### Generate some data
x = np.linspace(-10, 10, 200)
y = np.linspace(-10, 10, 200)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Create a figure and axes object
fig, ax = plt.subplots(figsize=(10,8))
# Plot the function with a colormap
cmap = plt.cm.Spectral
im = ax.imshow(Z, cmap=cmap, extent=(-10, 10, -10, 10))

# Add a colorbar for the colormap
cbar = fig.colorbar(im, ax=ax)
cbar.ax.set_ylabel('Function Value', fontsize=12)

# Add contour lines to the plot
levels = np.linspace(-0.5, 0.5, 21)
cont = ax.contour(X, Y, Z, levels=levels, colors='black', linewidths=0.5)
# Label for the contour 
ax.clabel(cont, fmt='%1.2f', fontsize=8)

# Add a title and axis labels
ax.set_title('Strange Function Plot', fontsize=14)
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)

# Add a grid to the plot
ax.grid(True, linestyle='--', linewidth=0.5)

# Add text annotations to the plot in a frame bbox_props
bbox_props = dict(boxstyle='round', facecolor='white', alpha=0.5)
ax.text(-9, 9, 'Ratio', fontsize=10, color='black', fontweight='bold',  bbox=bbox_props)
ax.text(3.5, -9, 'Ratio', fontsize=10, color='black',  fontweight='bold',  bbox=bbox_props)

## Add a green arrow to the plot
ax.arrow(5, 5, -5, -5, width=0.1, length_includes_head=True,
        head_width=0.3, head_length=0.5, color='green')

## Set the x and y-axis limits
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
```
