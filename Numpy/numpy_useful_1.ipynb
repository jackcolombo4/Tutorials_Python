<h3 style="color:moccasin "> => Recap: </h3>
<div style="margin-top: -8px;">
Ndim: <br>    
Number of dimensions (or axes) of the array => 2 if 2-dimensional array.       

Shape:      
The number of elements (size) in each dimension as a tuple.        
Dimension is the number of indices or subscripts, that we require in order to specify an individual element of an array.       
=> if dim=6 shape can be (1,2,3,4,5,6).            
=> arr4.shape is (3, 1).     

Dtype: <br>
Data type of the elements in the array (int64).      

Size:<br>
Total number of elements in the array.      
=> arr5 has 3 rows and 3 column, the total number of elements is 3*3 = 9.       

Itemsize: <br>
Size (in bytes) of each element in the array. For an int64 data type, each element takes up 8 bytes.      

Total_size: <br>     
arr4.size * arr4.itemsize: 
The total number of bytes based on the number of elements and the size of each element.       
For arr4, the calculation is 3 * 8 = 24       

Nbytes: <br>
Total number of bytes used by the array data.    
=> arr4 has 3 elements and each element is 8 bytes, the total size is 3 * 8 = 24 bytes        
=> arr5.nbytes = arr5.size * arr5.itemsize = 9 * 8 = 72 bytes  
</div>

```python
def is_reshape_valid(arr, new_shape):
    # Compute the size of the original array
    original_size = arr.size
    # Compute the size of the reshaped array
    new_size = np.prod(new_shape)
    # Check if the sizes match
    if original_size == new_size:
        return True
    else:
        return False

ret = is_reshape_valid(arr,(2,20))
print()
print(ret)
```

```python
def get_valid_reshape_shapes_with_minus_1(arr):
    '''The -1 is a special value that can be used for one dimension of the output shape to indicate that its size should be automatically 
    calculated in the 'reshape' based on the size of the input array and the other specified dimension(s).'''
    valid_shapes = []
    size = arr.size

    for i in range(1, size + 1):
        if size % i == 0:
            valid_shapes.append((-1, i))

    for i in range(2, int(np.sqrt(size)) + 1):
        if size % i == 0:
            valid_shapes.append((-1, i, size // i))
            valid_shapes.append((-1, size // i, i))

    return valid_shapes

retta = get_valid_reshape_shapes_with_minus_1(arr)
print()
print(retta)
```

```python
def get_valid_reshape_shapes(arr):
    valid_shapes = []
    size = arr.size

    for i in range(1, size + 1):
        if size % i == 0:
            valid_shapes.append((i, size // i))

    return valid_shapes

rere = get_valid_reshape_shapes(arr)
print()
# prova_res = arr.reshape(2,20)
# prova_res = arr.reshape(-1, 2, 20)
print(rere)
print()
print()
#print(prova_res)
```
